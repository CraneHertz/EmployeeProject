```python
import pandas as pd
import numpy as np

# 生成 50 个员工的随机 4 位数 ID
np.random.seed(42)
employee_ids = np.random.choice(range(1000, 10000), size=50, replace=False)

# 定义部门列表
departments = ['Administration', 'Sales', 'Development', 'Data Science', 'Research & Development']

# 随机分配部门给每个员工
current_departments = np.random.choice(departments, size=50, replace=True)

# 创建包含 50 个员工的初始数据框，并添加 'current_department' 列
data = {
    'employee_id': employee_ids,
    'Education Level': np.random.randint(1, 6, size=50),
    'Leadership': np.random.randint(1, 6, size=50),
    'Communication Skills': np.random.randint(1, 6, size=50),
    'Self-Motivation': np.random.randint(1, 6, size=50),
    'Problem-Solving Skills': np.random.randint(1, 6, size=50),
    'Technical Proficiency': np.random.randint(1, 6, size=50),
    'Emotional Intelligence': np.random.randint(1, 6, size=50),
    'Creativity': np.random.randint(1, 6, size=50),
    'Teamwork': np.random.randint(1, 6, size=50),
    'Decision-Making': np.random.randint(1, 6, size=50),
    'current_department': current_departments
}

df_selected = pd.DataFrame(data)

# 生成 12 个月的日期
months = pd.date_range(start='2023-01-01', periods=12, freq='M')

# 扩展数据以包含每个员工每个月的数据，并添加随机变化
monthly_data = []

for month in months:
    df_month = df_selected.copy()
    df_month['date'] = month

    # 对各表现指标添加随机波动
    for column in ['Leadership', 'Communication Skills', 'Self-Motivation',
                   'Problem-Solving Skills', 'Technical Proficiency',
                   'Emotional Intelligence', 'Creativity', 'Teamwork', 'Decision-Making']:
        df_month[column] = np.clip(df_month[column] + np.random.randint(-1, 2, size=len(df_month)), 1, 5)

    monthly_data.append(df_month)

# 合并所有月度数据
df_monthly = pd.concat(monthly_data, ignore_index=True)

# 打印前 24 行数据（用于检查）
print(df_monthly.head(24))




```

```python
# 输出到 CSV 文件
df_monthly.to_csv('employee_monthly_data.csv', index=False)
```
